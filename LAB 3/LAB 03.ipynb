```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

**ZADATAK 1: Otvoriti i prikazatu fotografiju giaraffes.jpg koja s enalazi u istom folderu kao i ipynb fajl.**

```python
slika = cv2.imread('giraffes.jpg')
slika_rgb = cv2.cvtColor(slika, cv2.COLOR_BGR2RGB)
```

```python
display_img(slika_rgb)
```

**ZADATAK 2: Primjenit binary threshold filter na otvorenu fotografiju.**

```python
slika2 = cv2.imread("giraffes.jpg", 0)
ret,slika_binary_tresh = cv2.threshold(slika2,127,255,cv2.THRESH_BINARY)
```

```python
display_img(slika_binary_tresh, cmap='gray')
```

**ZADATAK 3: Otvoriti fotografiju giaraffes.jpg i konvertovati njen colorspace (prostor boja) u HSV i prikazati fotografiju.**

```python
slika3 = cv2.imread('giraffes.jpg')
slika_hsv = cv2.cvtColor(slika3, cv2.COLOR_BGR2HSV)
```

```python
display_img(slika_hsv)
```

**ZADATAK 4: Kreirati filter sa 4 x 4 Kernel fispunjenim vrijednostima 1/10 (0.01) i onda koristeći 2-D Convolution fiter primjeniti blur efekat na giraffer fotografiju. Prikaz slike raditi u normalnom RGB prostoru boja**

```python
kernel = np.ones(shape=(4,4),dtype=np.float32)/10
```

```python
kernel
```

```python
slika_blur = cv2.filter2D(slika_rgb,-1,kernel)
```

```python
display_img(slika_blur)
```

**ZADATAK 5: Kreirati Horizontal Sobel Filter (sobelx sa vježbi) sa kernelom veličine 5 na crnobijelu verziju giaraffes.jpg fotografije i prikažite rezultujuću fotografiju.**

```python
slika4 = cv2.imread('giraffes.jpg', 0)
```

```python
sobelx = cv2.Sobel(slika4,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx, cmap='gray')
```

**ZADATAK 6: Krirajte color histogram za RED, BLUE, i GREEN kanale giaraffe.jpg fotografije.**

```python
slika5 = slika.copy()
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([slika5],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Giraffes Histograms')
plt.show()
```
