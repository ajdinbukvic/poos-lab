#### ZADATAK 1: Otvorite fotografiju *dog_backpack.jpg* i prikažite je u notebook-u. Koristite korektan redosliejd za RGB prostor boje.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
slika = cv2.imread("dog_backpack.jpg")
```

```python
plt.imshow(slika)
```

```python
slika_rgb = cv2.cvtColor(slika, cv2.COLOR_BGR2RGB)
plt.imshow(slika_rgb)
```

#### ZADATAK 2: Preokrenuti sliku naopako i prikazati je u notebook-u.

```python
slika_naopako = cv2.flip(slika_rgb, 0)
plt.imshow(slika_naopako)
```

#### ZADATAK 3: Iscrtati prazan CRVENI pravugaonik oko glave psa i prikazati sliku u notebook-u.

```python
nova_slika = slika_rgb.copy()
plt.imshow(nova_slika)
```

```python
cv2.rectangle(nova_slika,pt1=(200,350),pt2=(600,750),color=(255,0,0),thickness=5)
plt.imshow(nova_slika)
```

#### ZADATAK 4: Nacrtati plavi trougao u sredini slike. Smai odlučite o veličini u uglovima trougla.

```python
nova_slika2 = slika_rgb.copy()
height, width, channel = nova_slika2.shape
sredina = (width // 2, height // 2)
stranica = 600
point1 = (sredina[0], sredina[1] - int(np.sqrt(3) * stranica / 3))
point2 = (sredina[0] - stranica // 2, sredina[1] + int(np.sqrt(3) * stranica / 6))
point3 = (sredina[0] + stranica // 2, sredina[1] + int(np.sqrt(3) * stranica / 6))
cv2.line(nova_slika2,point1,point2,(0,0,255),6)
cv2.line(nova_slika2,point2,point3,(0,0,255),6)
cv2.line(nova_slika2,point3,point1,(0,0,255),6)
plt.imshow(nova_slika2)
```

### BONUS ZADATAK. Ispunite trougao bojom? Potrebna funkcija koju nismo do sada koristili. Istražiti link ispod:.

[OpenCV dokumentacija](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
pts = np.array([point1, point2, point3], np.int32)
pts = pts.reshape((-1, 1, 2))
cv2.fillPoly(nova_slika2, [pts], color=(0,0,255))
plt.imshow(nova_slika2)
```
